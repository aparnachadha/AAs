```
#Additional-Activity-5

#predefine code start
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('https://shravantihable.github.io/AA/inpe_brazilian_amazon_fires_1999_2019.csv')
df

group_by_year = df.groupby('year')['state'].count().reset_index()
group_by_year

sorted_df = group_by_year.sort_values(by=['state'], ascending=False)
sorted_df

#predefine code ends


```

```
#using print() function add the string as "In which year the maximum states where affected by fire and how many?"
print("In which year the maximum states where affected by fire and how many?")
#retrive the first row from the sorted_df data frame using the head() function.
print(sorted_df.head(1))

#using the print() function print the code for next line "\n"
print("\n")

#using print() function add the string as "In which year the mimimum states where affected by fire ?"
print("In which year the mimimum states where affected by fire ?")
#retrive the last row of the 'year' column from the sorted_df data frame using the tail() function.
print(sorted_df['year'].tail(1))

#using the print() function print the code for next line "\n"
print("\n")

#using print() function add the string as "How many states where affected by fire in year 2019 ?"
print("How many states where affected by fire in year 2019 ?")

#using sorted_df.loc[] function check the values of 'year' column of sorted_df data frame is equal to 2019 and print it using the print function.
print(sorted_df.loc[sorted_df['year'] == 2019])

#using the print() function print the code for next line "\n"
print("\n")
```
