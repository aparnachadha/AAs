```
#Additional-Activity 3
#plot a scatter plot showing the maximum values of a firespots in a year.

#predefine code
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('https://shravantihable.github.io/AA/inpe_brazilian_amazon_fires_1999_2019.csv')
df
#predefine code ends

#apply the groupby function on 'year' column and add a column  'firespots' , on it apply the max() function, and reset the index to 0 and store it in variable group_by_year .

group_by_year = df.groupby('year')['firespots'].max().reset_index()
#print the group_by_year
group_by_year


#using the plt.figure() function set the figure size of the plot by setting the width and height of the plot.
plt.figure(figsize=(12, 8))
#using the plt.xticks() function set the rotation as vertical to display the label names vertically on x-axis
plt.xticks(rotation='vertical')

#using the plt.scatter() function to plot the scatter plot and set the parameters in following way:
#group_by_year['year']
#group_by_year['firespots']
#c- set the color of your choice
# label= 'Firespots'
#alpha=0.5 - make the color of the bubble as solid
#s - set the size of the bubble as group_by_year['firespots']/10
plt.scatter(group_by_year['year'], group_by_year['firespots'], 
                  c='red',label = 'Firespots', 
                 alpha=0.5,
                 s = group_by_year['firespots']/10)

#predefine code
plt.legend()
plt.xlabel("State", size=14)
plt.ylabel("Firespots", size=14)
#predefine code end
```
