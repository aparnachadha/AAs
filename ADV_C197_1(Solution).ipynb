```
#Activity 1- to fetch the month and time from the data frame.

# Import all the libraries and read 2015_Air_quality_in_northern_Taiwan.csv
#Predefine Code start
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('https://shravantihable.github.io/AA/2015_Air_quality_in_northern_Taiwan.csv')
df
#Predefine code end

#Inside the dataframe df, create a column 'month' like this df['month'] and inside it store the fetched month from 'time' column. For this access the DatetimeIndex() function using the pd alias,and inside the function write df['time'], and using the '.' operator fetch the month from it.
df['month'] = pd.DatetimeIndex(df['time']).month
#Inside the dataframe df, create a column 'time-hours' like this df['time'] and inside it store the fetched time from 'time' column. For this access the DatetimeIndex() function using the pd alias,and inside the function write df['time'], and using the '.' operator fetch the time from it.
df['time_hours'] = pd.DatetimeIndex(df['time']).time
#print the variable df to get the output.
df


#Note: run the code, you will get the updated dataframe/table, with month and time_hours columns added to it.
```
