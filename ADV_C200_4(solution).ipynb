```
# Task description - Check the user selected value and based on it extract the value from time column
from ipywidgets import widgets
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('https://shravaripatil.github.io/ADV/AA/2015_Air_quality_in_northern_Taiwan.csv')
extract_from_date_column = ['Choose one.. ','month','year', 'day']

def print_dataframe(b):
    global df
    display(df)
    print('Select a date and time column. Note will give an error any other column other than date time column')
    column_widget = widgets.Dropdown(options = df.columns)
    select_widget = widgets.Dropdown(options = extract_from_date_column)
    widgets1 = widgets.interactive(add_new_column, column_name=column_widget, extract= select_widget)
    display(widgets1)

def add_new_column(column_name, extract):
    global df
    # Use if statement to check if extract variable holds string 'day' and inside this if block write code to access day from the 'time' column of the dataframe and store it in a new column of dataframe
        # To create a new column use df variable with [] and inside it give a string which will be the name of your new column and assign this new column ,column values. Use pd.DatetimeIndex() function and pass the column name to this function. and use the dot operator with 'day' attribute
     # Use elif statement to check if extract variable holds string 'year' and inside this if block write code to access year from the 'time' column of the dataframe and store it in a new column of dataframe
        # To create a new column use df variable with [] and inside it give a string which will be the name of your new column and assign this new column ,column values. Use pd.DatetimeIndex() function and pass the column name to this function. and use the dot operator with 'year' attribute
    # Use elif statement to check if extract variable holds string 'month' and inside this if block write code to access month from the 'time' column of the dataframe and store it in a new column of dataframe
        # To create a new column use df variable with [] and inside it give a string which will be the name of your new column and assign this new column ,column values. Use pd.DatetimeIndex() function and pass the column name to this function. and use the dot operator with 'month' attribute
    # Use the else block and inside the else block print 'Choose timestamp'
    if(extract == 'day'):
        df['date'] = pd.DatetimeIndex(df['time']).day
    elif(extract == 'year'):
        df['year'] = pd.DatetimeIndex(df['time']).year
    elif(extract == 'month'):
        df['month'] = pd.DatetimeIndex(df['time']).month
    else:
        print('Choose timestamp')
    # Print the dataframe using the display() function
    display(df)

show_dataframe= widgets.Button(description="Show dataframe")
show_dataframe.on_click(print_dataframe)
display(show_dataframe)
```
