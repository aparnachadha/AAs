```
# Task description - Write code to filter the data as per the user selected value on user selected dropdown
from ipywidgets import widgets
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('https://shravaripatil.github.io/ADV/AA/2015_Air_quality_in_northern_Taiwan.csv')
display(df)

input_box= widgets.Text()
display(input_box)

def print_dataframe(b):
  # to use the variable df inside this function define it as global variable
  global df
  #  we filter the dataframe by a specific value from a column
  # To filter the dataframe by the value entered by the user in the input field ,we enter the value in the input box based on which we want to filter the data so select the column on which we want to apply the filter and press the button
  # To get all the rows that satisfy a particular condition use df.loc[]
  #   condition - 
          # To access a column entered by the user we df [column_name] as the user selects the column name from a drop down. Therefore we will access the dropdown value inside df[] (Hint: to access the dropdown value we use variable which holds the dropdown widget(select_widget) with dot operator and 'value' attribute)
          # And then check if the column values is equal to the value entered by the user in the input box (Hint: To get the value entered by the user in input box user variable which holds the inputbox widget(input_box) with dot operator and 'value' attribute)
  # df.loc[] filters the dataframe as per the condition given inside it and creates a new dataframe. So then store the new dataframe in a variable
  new_df = df.loc[df[select_widget.value] == input_box.value]
  # Display the dataframe using display(). (Hint- Use display() and pass the variable holding new dataframe as parameter to display() function)
  display(new_df)

select_widget = widgets.Dropdown(options = df.columns)
display(select_widget)

show_dataframe= widgets.Button(description="Show dataframe")
show_dataframe.on_click(print_dataframe)
display(show_dataframe)
```
